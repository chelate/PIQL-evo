```julia
using Plots
using Statistics
include("Enviroment/gridworld.jl")
include("Enviroment/agent.jl")
```

```julia
#Defining a normal maze

size=(10,10) # set size of the maze
walls=[] # manually set walls into the maze
location_of_reward=(6,7) #x and y value of maze goal 
rewards=[-1,20] # first index is the cost of normal possisitons second is the reward of the goal



testgrid=initilaize_gridwold(size,walls,location_of_reward,rewards)
grid=copy(testgrid.grid)
heatmap((grid),fill=true,levels=[-2,-1,0,1,10],color=cgrad(:roma, [0.02, 0.4, 0.9,1], rev = false, categorical = true))

```

```julia
#Defining a  pickel maze
size=(15,15)
thicknes_of_pickle=5 
holes_in_pickle=15 # sets the number of holes in the pickle be careful not to build a strage pickle when this becomes to large  
location_of_reward=(13,12)
rewards=[-1,20]
testgrid=initilaize_gridwold_pickle(size, holes_in_pickle,thicknes_of_pickle,location_of_reward,rewards)

grid=copy(testgrid.grid)
heatmap((grid),fill=true,levels=[-2,-1,0,1,10],color=cgrad(:roma, [0.02, 0.4, 0.9,1], rev = false, categorical = true))

# please chack that maze is ok ... if not just run again
```

```julia
# Please run the cell of the QL you want to use
```

```julia
#define epsilon greedy QL 
include("Learners/SQL.jl")
epsilon=0.1 # for epsilon greedy
gamma= 0.9 # decay of reward
omega=0.5  # sets how fast q gets updated updats are scaled by N^(- omega) where N is number of updates of that Q value

ql=initilize_RL_system(testgrid,epsilon,gamma,omega);

```

```julia
#define a soft  QL 
include("Learners/SQL.jl")
gamma,omega=0.9,0.5
β,β_LR=1,0.01  # β is the initial beta/ β_LR is the increase of that temp per round
ql=initilize_RL_system(testgrid,β,gamma,β_LR,omega);
```

```julia
#define a soft  QL 
include("Learners/MIRL_QL.jl")
gamma,omega=0.9,0.5
β,β_LR=1,0.01
prior_LR=0.1 # learning rate of the prior
ql=initilize_RL_system(testgrid,β,gamma,β_LR,prior_LR,omega);
```

```julia
#define an actor 
agent1=start_agent(testgrid)
```

```julia
# do some initial training

rounds=1000
last_round=0
foundcounter =0
totalrards=Vector{Int64}()
roundcounter=Vector{Int64}()
for round in 1:rounds
    action=choose_action(ql,agent1)
    found,reward = move_actor(testgrid, agent1,action)
    update_System(ql,agent1,action,reward)
    if found
        foundcounter+=1
        append!(totalrards,agent1.totalreward)
        append!(roundcounter,round-last_round)
        if (foundcounter%4 == 0)
            println("mean time to find the goal for 4 instences was $(mean(roundcounter))  with mean reward $(mean(totalrards))")
            totalrards=Vector{Int64}()
            roundcounter=Vector{Int64}()
        end
        last_round=round
        agent1=start_agent(testgrid)
        
    end
end

```

```julia
# do some more  training

rounds=100000
last_round=0
foundcounter =0
totalrards=Vector{Int64}()
roundcounter=Vector{Int64}()
for round in 1:rounds
    action=choose_action(ql,agent1)
    found,reward = move_actor(testgrid, agent1,action)
    update_System(ql,agent1,action,reward)
    if found
        foundcounter+=1
        append!(totalrards,agent1.totalreward)
        append!(roundcounter,round-last_round)
        if (foundcounter%100 == 0)
            println("mean time to find the goal for 100 instences was $(mean(roundcounter))  with mean reward $(mean(totalrards))")
            totalrards=Vector{Int64}()
            roundcounter=Vector{Int64}()
        end
        last_round=round
        agent1=start_agent(testgrid)
        
    end
end
```
